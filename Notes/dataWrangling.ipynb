# Data Wrangling with Pandas

## Sorting

```python
import pandas as pd
mtcars = pd.read_csv("../Datasets/mtcars.csv")
```

```python
# Look at the mtcars data 
mtcars.head()
```

```python
# Sort the data
mtcars.sort_values(by="mpg", ascending=False, inplace=True)

# inplace=True means the sorted values replace the mtcars dataset. inplace=False returns the sorted data
mtcars.head()
```

```python
# Sort the dataframe by mpg and disp columns
mtcars.sort_values(by=["mpg", "disp"], ascending=[False, True], inplace=True)

mtcars.head()
```

```python
# Get the n largest values in a specified column
mtcars.nlargest(n=6, columns="mpg")
```

## Missing and Duplicate Data

```python
# Returns a dataframe of boolean values indicating whether the given cell is null
mtcars.isna()

# mtcars.isnull() is an alias for isna()

# returns the number of true values.
mtcars.isna().sum() 
```
